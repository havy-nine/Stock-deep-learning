```python
#필요한 모듈들을 import
import pandas as pd
import os
from bs4 import BeautifulSoup
from selenium import webdriver
import csv
import time

#약 1년간 구글 주식 데이터가 나와있는 사이트 열기
path = "c:/Temp/chromedriver_240/chromedriver.exe"
driver=webdriver.Chrome(path)
url = 'https://finance.yahoo.com/quote/GOOG/history/'
driver.get(url)
```

```python
#표를 끝까지 스크롤 다운, 무한 스크롤 다운을 사용해 봤는데 document.body.scrollHeight를
#표가 아닌 광고의 끝으로 인식하여 그냥 2번 스크롤 다운함. 
driver.execute_script('window.scrollTo(0, 10000)')
time.sleep(1)
driver.execute_script('window.scrollTo(0, 10000)')
```

```python
#표 부분의 row 를 찾아 리스트로 저장
depth_1_tbody = driver.find_element_by_xpath('//*[@id="Col1-1-HistoricalDataTable-Proxy"]/section/div[2]/table/tbody')
depth_2_tr = depth_1_tbody.find_elements_by_tag_name("tr")
```

```python
depth_2_tr[0]
len(depth_2_tr)
```

```python
depth_2_tr[0].text
```

```python
#각각의 colume 에 대한 리스트를 만들고 
date2=[]
open2=[]
high2=[]
low2=[]
close2=[]
Volume2=[]
indcr2=[]
dif_hl=[]
dif_se=[]
#row의 데이터를 하나씩 불러오며 알맞은 리스트에 저장
for td in depth_2_tr:
    row = td.text 
    row_list = row.split(' ')
    
    date = row_list[0]+row_list[1]+row_list[2]
    date2.append(date)
    #open에 들어갈것에 대해 실수로 변경
    row_o = row_list[3].split(',')
    frow_o = row_o[0]+ row_o[1]
    open2.append(float(frow_o))
    #high에 들어갈 것에 대해 실수로 변경
    row_h = row_list[4].split(',')
    frow_h = row_h[0]+ row_h[1]
    high2.append(float(frow_h))
    #low에 들어갈 것에 대해 실수로 변경
    row_l = row_list[5].split(',')
    frow_l = row_l[0]+ row_l[1]
    low2.append(float(frow_l))
    #close에 들어갈 것에 대해 실수로 변경
    row_c = row_list[6].split(',')
    frow_c = row_c[0]+ row_c[1]
    close2.append(float(frow_c))
    #bolume에 들어갈 것에 대해 실수로 변경
    row_v = row_list[8].split(',')
    try:
        frow_v = row_v[0]+ row_v[1]+ row_v[2]
    except IndexError:
        frow_v=row_v[0] + row_v[1]
    Volume2.append(float(frow_v))
    #주식의 가격이 증가하면 indcr2 리스트에 1 추가 감소하면 0 추가
    if float(frow_o) / float(frow_c) >= 1:
        indcr2.append('1')
    else:
        indcr2.append('0')
    #그날 최댓값과 최솟값의 차이
    dif_hl.append(float(frow_h)-float(frow_l))
    #시가와 종가의 차이
    if float(frow_c)-float(frow_o)>=0:
        dif_se.append(float(frow_c)-float(frow_o))
    else :
        dif_se.append(float(frow_o)-float(frow_c))
    
    print(row_list)
```

```python
#위의 데이터로 dataframe 만들기
date2.reverse()
open2.reverse()
high2.reverse()
low2.reverse()
close2.reverse()
Volume2.reverse()
dif_se.reverse()
dif_hl.reverse()


Google_stock = pd.DataFrame()
Google_stock['날짜']=date2
Google_stock['시가']=open2
Google_stock['고점']=high2
Google_stock['저점']=low2
Google_stock['종가']=close2
Google_stock['거래량']=Volume2
Google_stock['증감']=indcr2
Google_stock['종가-시가']=dif_se
Google_stock['최대-최소']=dif_hl
```

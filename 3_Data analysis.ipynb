```python
#시가와 종가에 따른 거래량 확인하기. 
```

```python
df_sev = df[['시가','종가', '거래량']].groupby(['시가'], as_index=False).mean().sort_values(by='시가', ascending=True)
```

```python
#한글 폰트 입력
font_path = "C:\\Windows\\Fonts\\HMKMMAG.TTF"
font_name = matplotlib.font_manager.FontProperties(fname=font_path).get_name()
plt.rc('font', family=font_name, size=14) 
#시가 종가 거래량의 pairplot 그려서 확인해보기

sns.pairplot(df_sev, kind = 'reg')  


plt.show()
```
